The data is known as Anscombe's Quartet and is used here to show the power of simple summary statisitics

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x1 = [10.0,8,13,9,11,14,6,4,12,7,5]
y1 = [8.04,6.95,7.58,8.81,8.33,9.96,7.24,4.26,10.84,4.82,5.68]

linear_model=np.polyfit(x1,y1,1)
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0,20)
plt.scatter(x1,y1)
plt.plot(x_s,linear_model_fn(x_s),color="green")

plt.show()
```

```python
x2 = [10.0,8,13,9,11,14,6,4,12,7,5]
y2 = [9.14,8.14,8.74,8.77,9.26,8.1,6.13,3.1,9.13,7.26,4.74]

linear_model=np.polyfit(x2,y2,1)
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0,20)
plt.scatter(x2,y2)
plt.plot(x_s,linear_model_fn(x_s),color="green")

plt.show()
```

```python
x3 = [10.0,8,13,9,11,14,6,4,12,7,5]
y3 = [7.46,6.77,12.74,7.11,7.81,8.84,6.08,5.39,8.15,6.42,5.73]

linear_model=np.polyfit(x3,y3,1)
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0,20)
plt.scatter(x3,y3)
plt.plot(x_s,linear_model_fn(x_s),color="green")

plt.show()
```

```python
x4 = [8,8,8,8,8,8,8,19,8,8,8]
y4 = [6.58,5.76,7.71,8.84,8.47,7.04,5.25,12.5,5.56,7.91,6.89]

linear_model=np.polyfit(x4,y4,1)
linear_model_fn=np.poly1d(linear_model)
x_s=np.arange(0,20)
plt.scatter(x4,y4)
plt.plot(x_s,linear_model_fn(x_s),color="green")

plt.show()
```

As seen by in the above graphics these data sets are fundamentally different but i will now show how some rudimentary statistical techniques can make them appear indiscernable.

# Mean

The mean is a commonly used average representing the central value of a set of numbers.

```python
def mean_of_x(x_values):
    mean = 0
    for x in x_values:
        mean +=x
    return mean/len(x_values)

print(mean_of_x(x1))
print(mean_of_x(x2))
print(mean_of_x(x3))
print(mean_of_x(x4))
```

as seen above the mean x value of all the graphs is 9.

```python
def mean_of_y(y_values):
    mean = 0
    for y in y_values:
        mean +=y
    return mean/len(y_values)

print(mean_of_y(y1))
print(mean_of_y(y2))
print(mean_of_y(y3))
print(mean_of_y(y4))
```

The mean y value is also approximatley the same across all four graphs.

# Correlation

Is a representation of how closely pairs of variables are linearly related, we'll be using numpy to calculate it rather than reinventing the wheel but you can double check if you like.

```python
def correlation(x_set,y_set):
    return np.corrcoef(x_set,y_set)[0,1]

print(correlation(x1,y1))
print(correlation(x2,y2))
print(correlation(x3,y3))
print(correlation(x4,y4))
```

<h2>Standard deviation and variance</h2>

Variance is a measure of the spread of data from its average value given by the the standard deviation squared which will again be the same for all 4 graphs.

```python
get_rho = lambda x:np.std(x)

print(get_rho(y1))
print(get_rho(y2))
print(get_rho(y3))
print(get_rho(y4))
```

You can just trust me that the x values also produce the same standard deviation dentoted by rho and thus the same sample variance.
